```python
'''
So, your whole process could consist of the 3 functions:
get_salary: check if the string has just one value of two values (by checking if it contains the symbol "-"), and then calls the appropriate function get_salary_value or get_salary_range
get_salary_value function returns a dictionary like {"value": 830, "type": "week"}
get_salary_range function returns a dictionary like {"min": 1000, "max": 2000, "type": "week"}
'''
```

```python
def get_salary_range(salary_string):
    '''
    return a range
    '''
    d=dict()
    l=salary_string.split(' ')
#     if l[0]=='Up':
#         return None
    Type=l[-1]
    Types=['hour','day','week','month','year'] # h='$18 - $30 an hour'
    if Type in Types:
        min= float(l[0].replace(',','').split('$')[1])
        max= float(l[2].replace(',','').split('$')[1])
    d['mean']=(min+max)/2
    d['type']=Type
    d['min']=min
    d['max']=max
    return d
```

```python
def get_salary_value(salary_string):
    '''
    return the value
    '''
    d=dict()
    l=salary_string.split(' ')
#     if l[0]=='Up':
#         return None
    Type=l[-1]
    Types=['hour','day','week','month','year']
    if Type in Types:
        #mean= float(l[0].split('$')[1].replace(',',''))
        mean= float(l[0].replace(',','').split('$')[1])
    d['mean']=mean
    d['type']=Type
    d['min']=None
    d['max']=None
    return d
```

```python
def get_salary(salary_string):
    print(salary_string)
    if '-' in salary_string:
        '''
        its a range
        '''
        return get_salary_range(salary_string)
    elif len(salary_string.split(' '))==3:
        '''
        only a value
        '''
        return get_salary_value(salary_string)
    else:
        return None
```

```python
def standardized(salary_string):
    if type(salary_string)!=str:
        return None
    if salary_string.strip()[0]!='$':
        return None
    d=get_salary(salary_string)
    if d==None:
        return None
    if d['type']=='hour':
        return d['mean']*40*52
    elif d['type']=='day':
        return d['mean']*5*52
    elif d['type']=='week':
        return d['mean']*52
    elif d['type']=='month':
        return d['mean']*12
    elif d['type']=='year':
        return d['mean']
    else:
        return None
```
